Je größer $w_1$ wird, desto schneller nimmt die Funktion $y_1(x)$ den Wert $1$ an. Die Funktion wird steiler.

#### b)
Die Funktion wird invertiert. D.h. sie startet bei $1$ und strebt dann gegen $0$.

#### c)
|                 | $w_1 > 0$ | $w_1 < 0$ |
|-----------------|-----------|-----------|
|$b_1 \uparrow$   | links     | rechts
|$b_1 \downarrow$ | rechts    | links

### 2.
- $w_1 = 1$ 
- $b_1 = -1$ 
- $w_2 = -1$ 
- $b_2 = -1$ 


```python
%matplotlib widget
from ipywidgets import interact
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1)

def f(x):
    return 1 / (1 + np.exp(-x))

def network(x, w1, b1, w2, b2):
    return f(w1 * x + b1) + f(w2 * x + b2)


def plot_network(w1, b1, w2, b2):
    values = []
    y1 = []
    y2 = []
    steps = np.arange(-5, 5, 0.1)
    for step in steps:
        values.append(network(step, w1, b1, w2, b2))
        y1.append(f(w1 * step + b1))
        y2.append(f(w2 * step + b2))

    ax.clear()
    ax.plot(steps, values, label="$y(x)$")
    ax.plot(steps, y1, label="$y_1(x)$")   
    ax.plot(steps, y2, label="$y_2(x)$")

    ax.grid(True)
    ax.set_xlabel("Time (s)")
    ax.legend(loc="lower right")

w1_slider = widgets.FloatSlider(min=-1, max=1, value=1, description='w1')
b1_slider = widgets.FloatSlider(min=-1, max=1, value=-1, description='b1')
w2_slider = widgets.FloatSlider(min=-1, max=1, value=-1, description='w2')
b2_slider = widgets.FloatSlider(min=-1, max=1, value=-1, description='b2')

interact(plot_network, w1=w1_slider, b1=b1_slider, w2=w2_slider, b2=b2_slider)
```
